```python
file  = './piano.jpg'
image = Image.open(file,'r')
```

```python
print(image)
```

```python
image.mode
```

```python
images = []
```

```python
from PIL import ImageEnhance
enhancer = ImageEnhance.Brightness(image)
```

```python
for i in range(10):
    images.append(enhancer.enhance(i/10))
    display(images[i])
```
